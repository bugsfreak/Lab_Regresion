## Preprocesamiento del dataset. Transformaciones previas necesarias para la modelación

```python
#Transformación de las caracteristicas 
#Libreria para el min max scaling
#Libreria para el split del train y test
#Libreria para la regresión lineal
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

## División del dataset en datos de entrenamiento y datos de test 

```python
#Código que realice la división en entrenamiento y test, de acuerdo con la estretgia de evaluación planeada. Describa cuál es.
X_train, X_test, y_train, y_test = train_test_split(dsEstates.drop('Y house price of unit area', axis=1),dsEstates['Y house price of unit area'],
                                                    test_size=0.3,
                                                    random_state=2)

X_train.shape, X_test.shape
```

```python
# Se instancia el escalador MixMax
escalador = MinMaxScaler()

# fit the scaler to the train set, it will learn the parameters
escalador.fit(X_train)

# transform train and test sets
X_train_escalado = escalador.transform(X_train)
X_test_escalado = escalador.transform(X_test)

X_train_escalado = pd.DataFrame(X_train_escalado, columns=X_train.columns)
X_test_escalado = pd.DataFrame(X_test_escalado, columns=X_test.columns)
```

```python
#El máximo dato escalado
escalador.data_max_
```

```python
#El mínimo dato escalado
escalador.min_
```

## Regresión Lineal

```python
#Códo del modelo
Rlineal = LinearRegression()
```

```python
#Entrenamiento
RL=Rlineal.fit(X_train_escalado,y_train)
```

```python
#Test
RLP =RL.predict(X_test)
```

## Evaluación del Modelo

Construya un o dos párrafos con los principales hallazgos. Cómo está funcionando el modelo? Calidad en los resultados de predicción. 

```python
# Medición de ajuste en datos de entrenamiento
print('Medición de ajuste:',metrics.r2_score(y_train, RL.predict(X_train_escalado)))
```

```python
Entrada = ['X1 transaction date','X2 house age','X3 distance to the nearest MRT station','X4 number of convenience stores','X5 latitude','X6 longitude']
Objetivo = ['Y house price of unit area']
TestingDataResults=pd.DataFrame(data=X_test, columns=Entrada)
TestingDataResults[Objetivo]=y_test
TestingDataResults[('Predicted'+'Price')]=np.round(RLP)

print(TestingDataResults[['Price','Predicted'+'Price']].head())
```
